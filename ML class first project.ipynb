```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
```

```python
x_1 = np.random.rand(5) * 20
x_2 = np.random.rand(25) * 20
x_3 = np.random.rand(100) * 20
x_4 = np.random.rand(1000) * 24
y_1 = 2.164 * x_1 - 3.656
y_1 = y_1 + np.random.normal(scale = 4, size = y_1.shape)
y_2 = 2.164 * x_2 - 3.656
y_2 = y_2 + np.random.normal(scale = 4, size = y_2.shape)
y_3 = 2.164 * x_3 - 3.656
y_3 = y_3 + np.random.normal(scale = 6, size = y_3.shape)
y_4 = 2.164 * x_4 - 3.656
y_4 = y_4 + np.random.normal(scale = 10, size = y_4.shape)
X_1 = x_1.reshape(-1, 1)
Y_1 = y_1.reshape(-1, 1)
X_2 = x_2.reshape(-1, 1)
Y_2 = y_2.reshape(-1, 1)
X_3 = x_3.reshape(-1, 1)
Y_3 = y_3.reshape(-1, 1)
X_4 = x_4.reshape(-1, 1)
Y_4 = y_4.reshape(-1, 1)




```

```python
y_index_1 = []
```

```python
lr_model = LinearRegression(normalize = True)
lr_model.fit(X_1, Y_1)
plt.plot(x_1, y_1, 'bo')
y1_line = lr_model.predict(X_1)
plt.plot(X_1,y1_line, 'r--')
y_index_1.append(float(lr_model.score(X_1, Y_1) ))
```

```python
lr_model.fit(X_2, Y_2)
plt.plot(x_2, y_2, 'bo')
y2_line = lr_model.predict(X_2)
plt.plot(X_2,y2_line, 'r--')
y_index_1.append(float(lr_model.score(X_2, Y_2) ))
```

```python
lr_model.fit(X_3, Y_3)
plt.plot(x_3, y_3, 'bo')
y3_line = lr_model.predict(X_3)
plt.plot(X_3,y3_line, 'r--')
y_index_1.append(float(lr_model.score(X_3, Y_3)))
```

```python
lr_model.fit(X_4, Y_4)
plt.plot(x_4, y_4, 'x' , color='blue')
y4_line = lr_model.predict(X_4)
plt.plot(X_4,y4_line, '--',color = 'maroon')
y_index_1.append(float(lr_model.score(X_4, Y_4) ))
```

```python
y_index_4 = []
```

```python
poly_reg_4 = PolynomialFeatures(degree = 4)
X_poly_4 = poly_reg_4.fit_transform(X_1)
lr_model.fit(X_poly_4,Y_1)
X_1_plot = np.arange(-10,30,0.1).reshape(-1,1)
X_1_plot_poly = poly_reg_4.fit_transform(X_1_plot)
plt.scatter(X_1, Y_1, color = "blue")
plt.plot(X_1_plot[:,0],lr_model.predict(X_1_plot_poly),'r--')
y_index_4.append(float(lr_model.score(X_poly_4,Y_1) ))
```

```python
poly_reg_4 = PolynomialFeatures(degree = 4)
X_poly_4 = poly_reg_4.fit_transform(X_2)
lr_model.fit(X_poly_4,Y_2)
X_2_plot = np.arange(-10,30,0.1).reshape(-1,1)
X_2_plot_poly = poly_reg_4.fit_transform(X_2_plot)
plt.scatter(X_2, Y_2, color = "blue")
plt.plot(X_2_plot[:,0],lr_model.predict(X_2_plot_poly),'r--')
y_index_4.append(float(lr_model.score(X_poly_4,Y_2) ))
```

```python
poly_reg_4 = PolynomialFeatures(degree = 4)
X_poly_4 = poly_reg_4.fit_transform(X_3)
lr_model.fit(X_poly_4,Y_3)
X_3_plot = np.arange(-10,30,0.1).reshape(-1,1)
X_3_plot_poly = poly_reg_4.fit_transform(X_3_plot)
plt.scatter(X_3, Y_3, color = "blue")
plt.plot(X_3_plot[:,0],lr_model.predict(X_3_plot_poly),'r--')
y_index_4.append(float(lr_model.score(X_poly_4,Y_3) ))
```

```python
poly_reg_4 = PolynomialFeatures(degree = 4)
X_poly_4 = poly_reg_4.fit_transform(X_4)
lr_model.fit(X_poly_4,Y_4)
X_4_plot = np.arange(-10,30,0.1).reshape(-1,1)
X_4_plot_poly = poly_reg_4.fit_transform(X_4_plot)
plt.scatter(X_4, Y_4, marker = 'x' , color = 'blue')
plt.plot(X_4_plot[:,0],lr_model.predict(X_4_plot_poly),'--',color = 'maroon')
y_index_4.append(float(lr_model.score(X_poly_4,Y_4) ))
```
